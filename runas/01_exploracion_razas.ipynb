## Observación importante — escala de valores

Las columnas numéricas clave del dataset AKC usan escala 0.2 - 1.0
mientras que Seiðr usa escala 0 - 5.

### Fórmula de conversión
`valor_seidr = round(valor_akc * 5)`

Ejemplos:
- 0.2 → 1
- 0.4 → 2
- 0.6 → 3
- 0.8 → 4
- 1.0 → 5

Esta conversión se aplicará en el bloque de limpieza de datos.

### Criterios reales para un perro de asistencia
Un buen perro de asistencia no es el más entrenable.
Es el que mejor encaja con el perfil específico de su usuario.

**Los criterios clave son:**
- Vínculo emocional con el usuario (demeanor_value)

- Adaptabilidad al entorno del usuario
- Energía compatible con el estilo de vida del usuario
- Tamaño adecuado para el contexto

**Ejemplos reales que lo confirman:**
- Spitz japonés: trainability moderada, vínculo excepcional, afectuoso
- Groenendael: inteligente, protectora, vínculo muy intenso con su persona

### Decisión para Seiðr
El algoritmo NO buscará la raza más entrenable.
Buscará la raza cuyo perfil de temperamento, energía y vínculo
encaje mejor con el perfil de 8 dimensiones del usuario.

## CORREGIMOS COSAS COMO RENOMBRAR COLUMNA 0 POR BREED O LA ESCALA DE PUNTUACIÓN

```python
# Convertir escala AKC (0.2-1.0) a escala Seiðr (0-5)
columnas_escala = [
    'grooming_frequency_value',
    'shedding_value', 
    'energy_level_value',
    'trainability_value',
    'demeanor_value'
]

for col in columnas_escala:  # recorremos cada columna numérica clave
    df[col + '_seidr'] = (df[col] * 5).round().astype('Int64')  # multiplicamos por 5 y redondeamos

# Verificar la conversión con un ejemplo
df[['breed', 'energy_level_value', 'energy_level_value_seidr', 
    'trainability_value', 'trainability_value_seidr']].head(10)
```

```python
# ============================================================
# EDA COMPLETO — Limpieza de datos
# ============================================================

# 1. Ver nulos actuales
print("=== NULOS POR COLUMNA ===")
print(df.isnull().sum())
print(f"\nTotal filas: {len(df)}")
print(f"Total columnas: {len(df.columns)}")
```

```python
# ============================================================
# EDA COMPLETO — Limpieza de datos
# ============================================================

# 1. Ver nulos actuales — cuántos valores faltan por columna
print("=== NULOS POR COLUMNA ===")
print(df.isnull().sum())  # suma los nulos de cada columna
print(f"\nTotal filas: {len(df)}")     # número de razas
print(f"Total columnas: {len(df.columns)}")  # número de características

# 2. Comprobar si hay filas duplicadas
print("\n=== LIMPIEZA ===")
print(f"Duplicados: {df.duplicated().sum()}")  # 0 = no hay duplicados, perfecto

# 3. Rellenar nulos en columnas clave con la mediana
# Usamos mediana (valor del medio) en lugar de media (promedio)
# porque la mediana no se ve afectada por valores extremos
columnas_clave = [
    'energy_level_value',       # nivel de energía en escala AKC (0.2-1.0)
    'trainability_value',       # capacidad de entrenamiento en escala AKC
    'demeanor_value',           # comportamiento/temperamento en escala AKC
    'energy_level_value_seidr', # nivel de energía convertido a escala Seiðr (0-5)
    'trainability_value_seidr', # entrenabilidad en escala Seiðr
    'demeanor_value_seidr'      # comportamiento en escala Seiðr
]

for col in columnas_clave:  # recorremos cada columna clave
    mediana = df[col].median()           # calculamos la mediana de esa columna
    nulos_antes = df[col].isnull().sum() # guardamos cuántos nulos había
    df[col] = df[col].fillna(mediana)    # rellenamos los nulos con la mediana
    print(f"{col}: {nulos_antes} nulos → rellenados con mediana ({mediana})")

# 4. Verificar que no quedan nulos en columnas clave
print(f"\nNulos restantes en columnas clave:")
print(df[columnas_clave].isnull().sum())  # debería ser 0 en todas
```

Confirmación del análisis anterior: el criterio de selección para Seiðr

adaptado al perfil del usuario.

```python
# ============================================================
# VISUALIZACIÓN 4 — Top 15 razas candidatas para asistencia
# ============================================================

# Creamos una puntuación compuesta para asistencia
# Combinamos entrenabilidad (40%) + vínculo (40%) + energía moderada (20%)
# La energía moderada se calcula como: 5 - abs(energy - 3) 
# Es decir, penalizamos las razas con energía muy alta o muy baja
# porque para asistencia lo ideal es energía media

df['puntuacion_asistencia'] = (
    df['trainability_value_seidr'] * 0.4 +   # 40% entrenabilidad
    df['demeanor_value_seidr'] * 0.4 +        # 40% vínculo/comportamiento
    (5 - abs(df['energy_level_value_seidr'] - 3)) * 0.2  # 20% energía moderada
)

# Ordenamos por puntuación y cogemos las 15 mejores
top15 = (df[['breed', 'puntuacion_asistencia', 
              'trainability_value_seidr', 
              'demeanor_value_seidr',
              'energy_level_value_seidr']]
         .sort_values('puntuacion_asistencia', ascending=False)  # orden descendente
         .head(15))  # las 15 primeras

# Gráfico de barras horizontal
fig4 = px.bar(
    top15,
    x='puntuacion_asistencia',   # eje X: puntuación total
    y='breed',                    # eje Y: nombre de la raza
    orientation='h',              # barras horizontales
    color='puntuacion_asistencia', # color según puntuación
    color_continuous_scale='Viridis',
    title='Top 15 razas candidatas para asistencia en Seiðr',
    text=top15['puntuacion_asistencia'].round(2)  # mostrar puntuación en la barra
)

fig4.update_layout(
    yaxis={'categoryorder': 'total ascending'},  # ordenar de menor a mayor visualmente
    xaxis_title='Puntuación de asistencia (0-5)',
    yaxis_title='',
    showlegend=False
)

fig4.show()

# Mostrar tabla con los datos completos
print("\nTop 15 razas candidatas:")
print(top15.to_string(index=False))
```

## Conclusión — Top 15 razas candidatas

El algoritmo de puntuación equilibra entrenabilidad (40%) + vínculo (40%) 
+ energía moderada (20%). El resultado es coherente con la realidad:
el Labrador Retriever aparece en el top, confirmando que el criterio funciona.

**Decisión de diseño importante:**
El tamaño NO es un factor penalizador en Seiðr.
Un perro de asistencia no tiene por qué ser grande.
Depende exclusivamente de las necesidades del usuario:
- Perro de alerta médica o apoyo emocional → tamaño irrelevante
- Perro guía → necesita tamaño adecuado
- Perro de servicio físico → necesita fuerza proporcional a la tarea

Seiðr filtrará por perfil de usuario, no por tamaño estándar.
