## Observación importante — escala de valores

Las columnas numéricas clave del dataset AKC usan escala 0.2 - 1.0
mientras que Seiðr usa escala 0 - 5.

### Fórmula de conversión
`valor_seidr = round(valor_akc * 5)`

Ejemplos:
- 0.2 → 1
- 0.4 → 2
- 0.6 → 3
- 0.8 → 4
- 1.0 → 5

Esta conversión se aplicará en el bloque de limpieza de datos.

## Razas candidatas para asistencia — primeras conclusiones

Las razas con trainability_value = 1.0 son fácilmente entrenables
pero eso no las convierte automáticamente en las mejores para asistencia.

### Criterios reales para un perro de asistencia
Un buen perro de asistencia no es el más entrenable.
Es el que mejor encaja con el perfil específico de su usuario.

**Los criterios clave son:**
- Vínculo emocional con el usuario (demeanor_value)
- Temperamento afectuoso y familiar (temperament)
- Adaptabilidad al entorno del usuario
- Energía compatible con el estilo de vida del usuario
- Tamaño adecuado para el contexto

**Ejemplos reales que lo confirman:**
- Spitz japonés: trainability moderada, vínculo excepcional, afectuoso
- Groenendael: inteligente, protectora, vínculo muy intenso con su persona

### Decisión para Seiðr
El algoritmo NO buscará la raza más entrenable.
Buscará la raza cuyo perfil de temperamento, energía y vínculo
encaje mejor con el perfil de 8 dimensiones del usuario.

```python
# Buscar razas específicas mencionadas como perros de asistencia con experiencia personal
razas_buscar = ['Spitz', 'Groenendael', 'Belgian', 'Malinois']

# Filtrar razas que contengan esos nombres
df[df['Unnamed: 0'].str.contains('|'.join(razas_buscar), case=False, na=False)][
    ['Unnamed: 0', 'temperament', 'energy_level_value', 'trainability_value', 'demeanor_value']
]
```

## Limitación del dataset AKC

El campo `demeanor_value` mide sociabilidad general con extraños,
no el vínculo con el usuario específico.

Un Groenendael tiene demeanor 0.4 (reservado con extraños)
pero vínculo excepcional con su persona de referencia.

### Decisión
Necesitamos añadir un campo `vinculo_usuario` en razas_asistencia.csv
que capture esta dimensión que el dataset AKC no mide.
Este campo se rellenará con investigación adicional de fuentes especializadas.

```python
# Renombrar columna sin nombre y limpiar nombres de columnas
df = df.rename(columns={'Unnamed: 0': 'breed'})  # renombramos a breed (raza en inglés)

# Verificar que el cambio se hizo bien
df.columns.tolist()  # muestra todos los nombres de columnas
```

## CORREGIMOS COSAS COMO RENOMBRAR COLUMNA 0 POR BREED O LA ESCALA DE PUNTUACIÓN

```python
# Convertir escala AKC (0.2-1.0) a escala Seiðr (0-5)
columnas_escala = [
    'grooming_frequency_value',
    'shedding_value', 
    'energy_level_value',
    'trainability_value',
    'demeanor_value'
]

for col in columnas_escala:  # recorremos cada columna numérica clave
    df[col + '_seidr'] = (df[col] * 5).round().astype('Int64')  # multiplicamos por 5 y redondeamos

# Verificar la conversión con un ejemplo
df[['breed', 'energy_level_value', 'energy_level_value_seidr', 
    'trainability_value', 'trainability_value_seidr']].head(10)
```

## Conversión de escala completada

Las columnas numéricas clave ahora tienen su equivalente en escala Seiðr (0-5):
- energy_level_value_seidr
- trainability_value_seidr
- demeanor_value_seidr
- grooming_frequency_value_seidr
- shedding_value_seidr
